```python
X = data.iloc[:,0:93]  # dataset excluding target attribute (encoded, one-hot-encoded,original)
Y = data[['Dos','normal','Probe','R2L','U2R']] # target attributes
```

```python
from sklearn.model_selection import train_test_split
from keras.layers import Dense
from keras.models import Sequential
from tensorflow.keras.layers import LeakyReLU

X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=42)
```

```python
from tensorflow.keras.activations import softmax

mlp = Sequential() # initializing model
# 1st
mlp.add(Dense(units=1650, input_dim=X_train.shape[1], ))
mlp.add(LeakyReLU())

#2nd hidden layer
mlp.add(Dense(units=1850))
mlp.add(LeakyReLU())
mlp.add(Dense(units=2048))
mlp.add(LeakyReLU())
mlp.add(Dense(units=1792))
mlp.add(LeakyReLU())
mlp.add(Dense(units=1536))
mlp.add(LeakyReLU())
mlp.add(Dense(units=1280))
mlp.add(LeakyReLU())
mlp.add(Dense(units=1024))
mlp.add(LeakyReLU())
mlp.add(Dense(units=768))
mlp.add(LeakyReLU())
mlp.add(Dense(units=512))
mlp.add(LeakyReLU())


# output layer with softmax activation
mlp.add(Dense(units=5,activation= 'softmax'))
```

```python
mlp.summary()
```

```python
from tensorflow.keras.optimizers import SGD
opt = SGD(learning_rate=0.001)
mlp.compile(loss="categorical_crossentropy",optimizer = opt , metrics = ['accuracy'])
```

```python
model = mlp.fit(X_train, y_train, epochs=1000, batch_size=5000, validation_split=0.2)
```
